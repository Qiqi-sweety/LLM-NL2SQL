```python
import matplotlib.pyplot as plt
import numpy as np

# 模型名称
models = ['CodeS 7B', 'Qwen 7B', 'Qwen2 7B', 'Code Qwen 7B', 'Symbol LLM 7B', 'LLaMA 8B', 'SQL Coder 8B', 
          'GLM 9B', 'Symbol LLM 13B', 'Qwen 14B', 'CodeS 15B', 'Qwen 32B', 'SQL Coder 34B']

# Spider数据集的性能（EM和EX）
spider_em = [34.3, 12.6, 33.7, 45.8, 50.2, 20.0, 4.1, 24.8, 58.9, 25.2, 36.0, 29.9, 20.0]
spider_ex = [64.4, 57.0, 70.5, 67.8, 65.2, 55.0, 57.3, 69.1, 72.8, 66.2, 66.3, 69.7, 57.9]

# 设置图表
fig, ax1 = plt.subplots(figsize=(12, 8))

# 设置柱状图位置
x = np.arange(len(models))
width = 0.35

# 绘制Spider数据集的柱状图
bar1 = ax1.bar(x - width/2, spider_em, width, label='Spider Exact Match Accuracy', color='blue')


# 创建第二个y轴
ax2 = ax1.twinx()

# 绘制VES值的折线图
bar2 = ax2.bar(x + width/2, spider_ex, width, label='Spider Execution Accuracy', color='orange')

# 图表设置
# ax1.set_xlabel('Models')
ax1.set_ylabel('Exact Match Accuracy (%)')
ax2.set_ylabel('Execution Accuracy (%)')
ax1.set_title('Performance Comparison on Spider Dataset')
ax1.set_xticks(x)
ax1.set_xticklabels(models, rotation=30, ha='right')

# 创建自定义图例条目
from matplotlib.patches import Patch
legend_elements = [Patch(facecolor='blue', label='Exact Match Accuracy'),
                   Patch(facecolor='orange', label='Execution Accuracy')]

# 添加图例到图表中
ax1.legend(handles=legend_elements, loc='upper left')

# 显示图表
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# 模型名称
models = ['CodeS 7B', 'Qwen 7B', 'Qwen2 7B', 'Code Qwen 7B', 'Symbol LLM 7B', 'LLaMA 8B', 'SQL Coder 8B', 
          'GLM 9B', 'Symbol LLM 13B', 'Qwen 14B', 'CodeS 15B', 'Qwen 32B', 'SQL Coder 34B']

# Bird数据集的性能（假设数据）
bird_accuracy = [17.28, 19.23, 33.90, 33.90, 15.25, 10.76, 34.09, 38.01, 19.36, 30.05, 22.29, 38.40, 35.53]
bird_ves = [18.84, 20.17, 34.37, 34.08, 17.77, 10.86, 35.01, 39.66, 20.21, 30.86, 22.97, 39.43, 38.30]

# 创建图表
fig, ax2 = plt.subplots(figsize=(12, 8))
ax3 = ax2.twinx()

# 设置图表位置
x = np.arange(len(models))
width = 0.35

# 绘制Bird数据集的柱状图和折线图
bar2 = ax2.bar(x, bird_accuracy, width, label='Accuracy', color='green')
line2 = ax3.plot(x, bird_ves, label='VES', color='red', marker='o', linestyle='dashed', linewidth=2)

# 设置Bird图表
ax2.set_ylabel('Accuracy (%)')
ax2.set_title('Performance on Bird Dataset')
ax2.set_xticks(x)
ax2.set_xticklabels(models, rotation=30, ha='right')
ax3.set_ylabel('VES')

# 创建自定义图例条目
from matplotlib.patches import Patch
from matplotlib.lines import Line2D
legend_elements = [Patch(facecolor='green', label='Accuracy'),
                   Line2D([0], [0], color='red', marker='o', linestyle='dashed', label='VES')]

# 添加图例到图表中
ax2.legend(handles=legend_elements, loc='upper left')

# 调整布局
plt.tight_layout()

# 显示图表
plt.show()
```
